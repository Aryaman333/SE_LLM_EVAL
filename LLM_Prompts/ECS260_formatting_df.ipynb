```python
import pandas as pd


responses = pd.read_excel('response.xlsx')
llm = pd.read_excel('llm.xlsx')

responses_map = {}
questions = responses.columns[5:]

for index, rows in responses.iterrows():
    for qn in questions:
        id = (rows['Age'], rows['Gender'], rows['Experience'], qn)
        if id in responses_map and rows[qn] in responses_map[id]:
            responses_map[id][rows[qn]] += 1
        else:
            if id not in responses_map:
                responses_map[id] = {}
            responses_map[id][rows[qn]] = 1

llm_map = {}
for index, rows in llm.iterrows():
    id = (rows['Age'], rows['Gender'], rows['Experience'], rows['Question'])
    answers = str(rows['answer']).split('\n')
    for answer in answers:
        if len(answer.split(':')) > 1:
            correct_ans = answer.split(':')[1].strip()
            if id in llm_map and correct_ans in llm_map[id]:
                llm_map[id][correct_ans] += 1
            else:
                if id not in llm_map:
                    llm_map[id] = {}
                llm_map[id][correct_ans] = 1

response_df = pd.DataFrame(columns=['Age', 'Gender', 'Experience', 'Question', 'Option', 'Count'])
for key, value in responses_map.items():
    for value_item in list(value.keys()):
        row = [key[0], key[1], key[2], key[3], value_item, value[value_item]]
        response_df. loc[len(response_df)] = row

llm_df = pd.DataFrame(columns=['Age', 'Gender', 'Experience', 'Question', 'Option', 'Count'])
for key, value in llm_map.items():
    for value_item in list(value.keys()):
        row = [key[0], key[1], key[2], key[3], value_item, value[value_item]]
        llm_df. loc[len(llm_df)] = row

llm_df.to_excel('llm_finals.xlsx')


```
